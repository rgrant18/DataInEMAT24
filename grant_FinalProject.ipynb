## The analysis I began to do on these datasets would be trying to see if there was a visible relationship between the two by just looking at the result of the graph below. I have learned that there is a positive correlation suggested by the two datasets. This would mean that the two datasets likely have an influence on each other.
```python
# Merging my two datasets in order to present them both in a matplotlib graph
aligned_data = pd.merge(dataset2, dataset1, left_index=True, right_index=True)
#Assigning the x and y values to their own variables in order to fill the scatterplot properly
x_values = aligned_data['Average (PPM)']
y_values = aligned_data['Annual Mean']
```

```python
# Creates our scatterplot
fig1, ax1 = plt.subplots()
ax1.scatter(x_values, y_values, alpha = .3)
ax1.set_xlabel("Time (Years)") #For some reason, I could not g
plt.show()
```
