```python
# Merging my two datasets in order to present them both in a matplotlib graph
aligned_data = pd.merge(dataset2, dataset1, left_index=True, right_index=True)
#Assigning the x and y values to their own variables in order to fill the scatterplot properly
x_values = aligned_data['Average (PPM)']
y_values = aligned_data['Annual Mean']
```

```python
fig1, ax1 = plt.subplots()
ax1.scatter(x_values, y_values, alpha = .3)
plt.show()
```

## In conclusion, with the data represented, there is a positive correlation between the two datasets.
